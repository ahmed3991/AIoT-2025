```python
# ============================================================
# ğŸ§  Practical Work: Deploying TinyML Models with TensorFlow Lite
# Dataset: Fashion-MNIST
# Target Device: Seeed Studio XIAO ESP32S3
# ============================================================

import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Conv2D, MaxPooling2D
import numpy as np
import os

# ============================================================
# ğŸ§© Task 1.1: Setup and Data Loading
# ============================================================

# Load the Fashion-MNIST dataset
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Normalize the image data to [0, 1]
train_images = train_images / 255.0
test_images = test_images / 255.0

# Prepare data for MLP model (flattened)
train_images_mlp = train_images.reshape((train_images.shape[0], 28 * 28))
test_images_mlp = test_images.reshape((test_images.shape[0], 28 * 28))

# Prepare data for CNN model (add channel dimension)
train_images_cnn = np.expand_dims(train_images, -1)
test_images_cnn = np.expand_dims(test_images, -1)

print("âœ… Data loaded successfully")
print("MLP Input Shape:", train_images_mlp.shape)
print("CNN Input Shape:", train_images_cnn.shape)

# ============================================================
# ğŸ§  Task 1.2: Define and Train MLP Model
# ============================================================

mlp_model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

mlp_model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

print("\nğŸ“˜ MLP Model Summary:")
mlp_model.summary()

print("\nğŸš€ Training MLP model...")
mlp_history = mlp_model.fit(
    train_images,
    train_labels,
    epochs=5,
    batch_size=64,
    validation_split=0.1,
    verbose=2
)

# ============================================================
# ğŸ§  Task 1.3: Define and Train CNN Model
# ============================================================

cnn_model = Sequential([
    Conv2D(16, kernel_size=3, activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=2),
    Conv2D(32, kernel_size=3, activation='relu'),
    MaxPooling2D(pool_size=2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

cnn_model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

print("\nğŸ“˜ CNN Model Summary:")
cnn_model.summary()

print("\nğŸš€ Training CNN model...")
cnn_history = cnn_model.fit(
    train_images_cnn,
    train_labels,
    epochs=5,
    batch_size=64,
    validation_split=0.1,
    verbose=2
)

# ============================================================
# ğŸ§ª Task 2.1: Evaluate Models
# ============================================================

print("\nğŸ“Š Evaluating MLP model...")
mlp_loss, mlp_accuracy = mlp_model.evaluate(test_images, test_labels, verbose=0)

print("\nğŸ“Š Evaluating CNN model...")
cnn_loss, cnn_accuracy = cnn_model.evaluate(test_images_cnn, test_labels, verbose=0)

print("\nâœ… --- Final Test Results ---")
print(f"MLP Model - Test Loss: {mlp_loss:.4f}, Test Accuracy: {mlp_accuracy:.4f}")
print(f"CNN Model - Test Loss: {cnn_loss:.4f}, Test Accuracy: {cnn_accuracy:.4f}")

# ============================================================
# ğŸ’¾ Task 2.2: Save Models and Get Sizes
# ============================================================

mlp_model.save('mlp_model.h5')
cnn_model.save('cnn_model.h5')

mlp_model_size = os.path.getsize('mlp_model.h5') / (1024 * 1024)
cnn_model_size = os.path.getsize('cnn_model.h5') / (1024 * 1024)

print("\nğŸ’¾ Model Files Saved:")
print(f"MLP model size: {mlp_model_size:.2f} MB")
print(f"CNN model size: {cnn_model_size:.2f} MB")


```

```python
# ============================================================
# âš™ï¸ Task 3.1: Convert and Quantize MLP Model (Full Integer Quantization)
# ============================================================

def representative_data_gen_mlp():
    for i in range(100):
        yield [train_images[i:i+1].astype(np.float32)]

converter = tf.lite.TFLiteConverter.from_keras_model(mlp_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.representative_dataset = representative_data_gen_mlp
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8

print("\nğŸ”„ Converting and quantizing MLP model...")
tflite_mlp_quant = converter.convert()
with open("mlp_model_quant.tflite", "wb") as f:
    f.write(tflite_mlp_quant)
print("âœ… Quantized MLP model saved as 'mlp_model_quant.tflite'")

# ============================================================
# âš™ï¸ Task 3.2: Convert and Quantize CNN Model (Full Integer Quantization)
# ============================================================

def representative_data_gen_cnn():
    for i in range(100):
        yield [train_images_cnn[i:i+1].astype(np.float32)]

converter = tf.lite.TFLiteConverter.from_keras_model(cnn_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.representative_dataset = representative_data_gen_cnn
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8

print("\nğŸ”„ Converting and quantizing CNN model...")
tflite_cnn_quant = converter.convert()
with open("cnn_model_quant.tflite", "wb") as f:
    f.write(tflite_cnn_quant)
print("âœ… Quantized CNN model saved as 'cnn_model_quant.tflite'")

# ============================================================
# ğŸ“ Task 4.1: Compare File Sizes
# ============================================================

mlp_tflite_size = os.path.getsize("mlp_model_quant.tflite") / (1024 * 1024)
cnn_tflite_size = os.path.getsize("cnn_model_quant.tflite") / (1024 * 1024)

print("\nğŸ“Š --- Model Size Comparison ---")
print(f"MLP original (.h5): {mlp_model_size:.2f} MB")
print(f"MLP quantized (.tflite): {mlp_tflite_size:.2f} MB")
print(f"CNN original (.h5): {cnn_model_size:.2f} MB")
print(f"CNN quantized (.tflite): {cnn_tflite_size:.2f} MB")

# ============================================================
# ğŸ§¾ Task 4.2: Deployment Feasibility Summary
# ============================================================

print("\nğŸ’¡ --- Deployment Feasibility (Seeed Studio XIAO ESP32S3) ---")
print(f"Available SRAM: 512 KB (â‰ˆ0.5 MB)\n")

if mlp_tflite_size < 0.3:
    print("âœ… MLP Model: Fits comfortably in SRAM.")
else:
    print("âŒ MLP Model: Too large for SRAM.")

if cnn_tflite_size < 0.3:
    print("âœ… CNN Model: Fits comfortably in SRAM.")
else:
    print("âš ï¸ CNN Model: Close to SRAM limit â€” may need pruning or smaller input size.")

print("\nğŸš€ Practical complete! Your TinyML models are ready for ESP32S3 deployment.")
```
